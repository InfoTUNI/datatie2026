## Mitä on data?

<img src="image/DIKW_Pyramid.svg.png" alt="DIKW pyramid: Data, Information, Knowledge, Wisdom" width="400">

*DIKW-pyramidi (Data–Information–Knowledge–Wisdom).*  
Kuva: Longlivetheux, Wikimedia Commons, lisenssi **CC BY-SA 4.0**.  
Lähde: https://commons.wikimedia.org/wiki/File:DIKW_Pyramid.svg

DIKW-pyramidi kuvaa usein käytettyä mallia, jossa **data, informaatio, tieto ja viisaus** nähdään hierarkkisina tasoina. Mallissa data muodostaa perustan, josta jalostetaan ensin informaatiota, sitten tietoa ja lopulta viisautta. 

- **Data** – käsittelemättömiä havaintoja tai faktoja  
- **Informaatio** – kontekstiin asetettua ja jäsenneltyä dataa  
- **Tieto (knowledge)** – ymmärrystä, joka syntyy informaation tulkinnasta  
- **Viisaus (wisdom)** – kyky käyttää tietoa harkitusti päätöksenteossa  

Ilkka Tuomi ([1999](https://doi.org/10.1080/07421222.1999.11518258)) käänsi aikoinaan pyramidin ylösalaisin: kaiken lähtökohta on tietämys, joka ohjaa datan tuottamista. Raakadataa ei ole olemassa, kiteyttää Gitelman ([2013](https://direct.mit.edu/books/edited-volume/3992/Raw-Data-Is-an-Oxymoron)). Alaimo ja Kallinikos ([2024](https://doi.org/10.7551/mitpress/11751.001.0001)) laajentavat keskustelun tietojohtamisen kontekstiin.

<img src="image/inverse-pyramid-tuomi.png" alt="The inverse pyramid (Tuomi)" width="400">

Datan tutkimus on, ehkä yllättävästikin, aktiivinen tutkimusalue. [Data Studies Bibliography](https://www.datastudiesbibliography.org/) kokoaa yhteen alan tutkimusta.

# Mitä on datatiede?

## Määritelmä

Datatieteilijän (data scientist) rooli organisaatiossa on moninainen. Työtä on kuvattu monitieteiseksi, yhdistäen ainakin tietoteknistä, matemaattista ja liiketoiminnallista osaamista. Harvard Business Review nosti ammattinimikkeen laajempaan tietoisuuteen vuonna 2012 artikkelillaan [Data Scientist: The Sexiest Job of the 21st Century](https://hbr.org/2012/10/data-scientist-the-sexiest-job-of-the-21st-century):

<blockquote>
"...he started to see possibilities. He began forming theories, testing hunches, and finding patterns that allowed him to predict whose networks a given profile would land in. He could imagine that new features capitalizing on the heuristics he was developing might provide value to users."
</blockquote>

Alunperin datatieteestä käytettiin termiä [datalogy](https://dl.acm.org/citation.cfm?id=366510). Mielenkiintoista luettavaa on esimerkiksi artikkeli [Datalogy — The copenhagen tradition of computer science](https://link.springer.com/article/10.1007/BF01941128) datalogy-termin kehittäjän Peter Naurin luomasta Kööpenhaminalaisesta tietotekniikan koulutuksen traditiosta. 

Datatiedettä voi tarkastella useilla eri painotuksilla. Se voidaan nähdä markkinointiterminä [tilastotieteelle](http://www2.isye.gatech.edu/~jeffwu/presentations/datascience.pdf). On kuitenkin selvää, että tämä hyvin kapea käsitys ei kuvaa datatiedettä riittävällä tavalla, vaan tilastotiede on nähtävä yhtenä osana [datatieteilijän osaamista](https://arxiv.org/ftp/arxiv/papers/1410/1410.3127.pdf). Simon Lindgren ([2020](https://www.wiley.com/en-us/Data+Theory%3A+Interpretive+Sociology+and+Computational+Methods-p-9781509539277)) ehdottaa, että datatiede voidaan myös tulkita erääksi tavaksi tehdä etnografista tutkimusta (ks. [Tikka et al., 2021](https://journal.fi/mediaviestinta/article/view/109871)).

Datatieteessä keskeistä on kyky muokata laajoja aineistoja sekä hyödyntää ohjelmointia eri tavoin. Esimerkkinä voidaan pitää siirtymää tilasto-ohjelmista kuten [SPSS](https://spss.fi/) tai [R](https://fi.wikipedia.org/wiki/R_(ohjelmointikieli)) kohti täysiverisiä ohjelmointikieliä kuten [Python](https://fi.wikipedia.org/wiki/Python_(ohjelmointikieli)). Sekä R että Python ovat käytännössä ohjelmointikieliä, mutta R erikoistuu nimenomaisesti tilastolliseen laskentaa ja grafiikan tuottamiseen. Mikä on siis muutos, joka on tapahtunut kun Python kasvattaa suosiotaan osin R:n kustannuksella?
