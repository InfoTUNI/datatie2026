#### Question 4 - Write code to extract age of soldiers from the date of birth column

```python
import datetime as dt
today = dt.datetime.today()
```

```python
df['Birth Date'] = pd.to_datetime(df['Birth Date'])
```

```python
def dob2age(dob):
    age = today - dob
    age = age.days/365
    return round(age)
```

```python
df['Age'] = df['Birth Date'].apply(dob2age)
df
```

#### Question 6 - write code to create a categorical attribute for Age
- if age <34 = low
- if age is b/w 34 to 40 - medium
- if age>40 - high

```python
df['Age_Cat_Bin'] = pd.cut(df['Age'],bins=[0,34,40,100],labels=['Low','Medium','High'])
df
```

```python
def age_category(age):
    if pd.isna(age):
        return 'Unknown'
    elif age <= 34:
        return 'Low'
    elif age <= 40:
        return 'Medium'
    else:
        return 'High'

df['Age_Cat_Func'] = df['Age'].apply(age_category)
```
